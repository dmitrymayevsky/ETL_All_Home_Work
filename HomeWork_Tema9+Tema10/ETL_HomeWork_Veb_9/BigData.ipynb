```scala
%pyspark

from pyspark.sql import Row
from pyspark.sql.functions import col, when, avg, round as spark_round

# Тестовые данные сгенерированы для этого задания специально, они представляют продажи интернет-магазина
sales_data = [
    Row(order_id=101, product="Laptop", category="Electronics", price=1200, quantity=1, customer="Alice", region="North", discount=0.05),
    Row(order_id=102, product="Mouse", category="Electronics", price=25, quantity=2, customer="Bob", region="South", discount=0.0),
    Row(order_id=103, product="Desk", category="Furniture", price=350, quantity=1, customer="Alice", region="North", discount=0.1),
    Row(order_id=104, product="Chair", category="Furniture", price=150, quantity=4, customer="Charlie", region="East", discount=0.0),
    Row(order_id=105, product="Monitor", category="Electronics", price=400, quantity=1, customer="Diana", region="West", discount=0.05),
    Row(order_id=106, product="Keyboard", category="Electronics", price=45, quantity=3, customer="Bob", region="South", discount=0.0),
    Row(order_id=107, product="Lamp", category="Furniture", price=30, quantity=2, customer="Eve", region="East", discount=0.0),
    Row(order_id=108, product="Laptop", category="Electronics", price=1200, quantity=1, customer="Charlie", region="East", discount=0.1),
]

df = spark.createDataFrame(sales_data)

print("=" * 60)
print("Продажи")
print("=" * 60)
df.show(10, truncate=False)

df_with_total = df.withColumn("total_amount", spark_round(col("price") * col("quantity") * (1 - col("discount")), 2))

print("\n" + "=" * 60)
print("Трансформация (пример 1): Фильтр (total_amount > 200)")
print("=" * 60)
high_value_orders = df_with_total.filter(col("total_amount") > 200)
high_value_orders.show(10, truncate=False)

print("\n" + "=" * 60)
print("Трансформация (пример 2): Агрегация - сумма продаж по категориям")
print("=" * 60)
from pyspark.sql.functions import sum as spark_sum
category_sales = df_with_total.groupBy("category").agg(spark_sum("total_amount").alias("total_revenue"))
category_sales.show(truncate=False)

print("\n" + "=" * 60)
print("Трансформация (пример 3): Категория цен (High если price > 300, иначе Low)")
print("=" * 60)
df_with_price_category = df_with_total.withColumn("price_category", when(col("price") > 300, "High").otherwise("Low"))
df_with_price_category.select("product", "price", "price_category").show(10, truncate=False)

print("\n" + "=" * 60)
print("Трансформация (пример 4): Статистика продаж по регионам")
print("=" * 60)
region_stats = df_with_total.groupBy("region").agg(
    spark_sum("total_amount").alias("total_sales"),
    avg("quantity").alias("avg_quantity")
)
region_stats.show(truncate=False)

print("\nОперация завершилась успешно")
```
